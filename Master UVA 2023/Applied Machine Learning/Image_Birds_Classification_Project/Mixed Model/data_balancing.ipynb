```python
import os
import shutil

def ensure_27_files(directory):
    # List all files in the directory
    files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]

    # Calculate the difference between the current file count and the desired count (27)
    diff = len(files) - 27

    if diff > 0:
        # If there are more than 27 files, remove the excess files
        files_to_remove = files[:diff]
        for file_to_remove in files_to_remove:
            file_path = os.path.join(directory, file_to_remove)
            os.remove(file_path)
            print(f"Removed: {file_path}")

    elif diff < 0:
        # If there are fewer than 27 files, create additional empty files
        files_to_create = ['empty_file_{}.txt'.format(i) for i in range(-diff)]
        for file_to_create in files_to_create:
            file_path = os.path.join(directory, file_to_create)
            with open(file_path, 'w'):
                pass
            print(f"Created: {file_path}")

if __name__ == "__main__":
    root_directory = './BirdDeep-2/train'  # Replace with the path to your root directory

    # Recursively traverse through subdirectories
    for subdir, _, _ in os.walk(root_directory):
        ensure_27_files(subdir)
```
